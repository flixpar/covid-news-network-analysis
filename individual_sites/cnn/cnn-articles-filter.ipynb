```python
import pandas as pd
```

```python
articles = pd.read_csv("data/cnn_articles_meta.csv")
```

```python
article_names = articles.title.to_list()
```

```python
keywords = ["covid", "coronavirus", "pandemic", "CDC"]
```

```python
relevant_titles = [t for t in article_names if any([k in t.lower() for k in keywords])]
relevant_titles
```
