```python
import csv
import pandas as pd
import requests
from bs4 import BeautifulSoup
import tqdm
```

```python
def parse_sitemap(url):
    page = requests.get(url,timeout = 5)
    soup = BeautifulSoup(page.content, "html.parser")
    links = [a["href"] for a in soup.select('a[href]') if '://'in a["href"] ]
    outside_links = [a for a in links if a.split('.')[1] != url.split('.')[1]]           
    return outside_links
```

```python
def geturl(a):
    d = pd.read_csv(a, usecols=['news_url'])
    url_list = d.values.tolist()
    return url_list
geturl('D:\data\csv\Real0501.csv')
```

```python
url_list = geturl('D:\data\csv\Real0501.csv')
outside_links0501 = []
for i in tqdm.tqdm(url_list):
    outside_links0501.append(parse_sitemap(i[0]))
df = pd.read_csv('D:\data\csv\Real0501.csv')
d = df.drop(["type","publish_date"],axis = 1)
d['links'] = outside_links0501
d.to_csv('D:\data\Real0501.csv')
```
