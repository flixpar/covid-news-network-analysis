```python
urlreg = "(https?[^\s]+)"
```

```python
tweetlink_reg = "…\s(https?[^\s]+)"
```

```python
outlinks_reg = "(?<!…\s)(https?[^\s]+)"
```

```python
tweet_links_raw = {}
for i, tweet in tqdm.tqdm(data.iterrows(), total=len(data)):
    text = tweet.text
    links = re.findall(outlinks_reg, text)
    if links:
        tweet_links_raw[i] = links
```

```python
tweet_links_raw
```

```python
len([u for v in tweet_links_raw.values() for u in v])
```

```python
with open("data/tweet_linkurls_raw.json", "w") as f:
    json.dump(tweet_links_raw, f)
```
