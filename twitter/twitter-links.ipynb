```python
import pandas as pd
import json
import tldextract
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
def url_to_domain(url):
    domain = tldextract.extract(url)
    domain = domain.domain + "." + domain.suffix
    return domain
```

```python
newsdata = pd.read_csv("../data/articles_collect/aylien_domains.csv")
source_domains = sorted(list(set(newsdata.domain.tolist())))
print(source_domains)
```

```python
tweetdata = pd.read_csv("../datasets/covid19_tweets.csv")
```

```python
with open("../data/twitter/tweet_linkurls_expanded.json", "r") as f:
    links_data = json.load(f)
```

```python
tweetdata["links"] = [links_data[str(i)] if str(i) in links_data else [] for i in range(len(tweetdata))]
```

```python
all_links = [u for l in links_data.values() for u in l]
all_domains = [url_to_domain(url) for url in all_links]
```

```python
tweet_domains = sorted(list(set(all_domains)))
print(tweet_domains)
```

```python
news_links = [all_links[i] for i in range(len(all_links)) if all_domains[i] in source_domains]
print(len(news_links))
print(news_links)
```

```python
news_domains_count = {d: all_domains.count(d) for d in source_domains}
print(news_domains_count)
```
